```python
from itertools import product

# Define the target sum
target_sum = 100

# Create a list to store the permutations
permutations_list = []

# Generate all permutations of 5 numbers from 1 to 100 with replacement
# Opinions are all multiples of 5 to reduce the size of the population
counter = 0
for perm in product(range(0, 101, ), repeat=5):
    if sum(perm) == target_sum:
        permutations_list.append(perm)

# Print the list of the first 1000 permutations
permutations_list[0:1000]

```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
matplotlib.use("TkAgg")

# Create a figure and axis for the animation
fig, ax = plt.subplots()

# Function to update the histogram in each frame
def update(frame):
    ax.clear()
    data = permutations_list[frame]
    ax.bar([1,2,3,4,5],data, alpha=0.75)
    ax.set_title(f'Permutation {frame + 1}')
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
    ax.set_xlim(-0.5, 6)

# Create an animation object and assign it to a variable
ani = FuncAnimation(fig, update, frames=len(permutations_list), repeat=False)

# Display the animation

```

```python
#Reduce the size of the population by implementing 
#Systematic Sampling by selecting every 15th element

filtered_list = []

for i in range(0,len(permutations_list)):
    if i%15 == 0:
        filtered_list.append(permutations_list[i])
```
